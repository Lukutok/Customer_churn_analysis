В датафрейме имеется 73 записи, повторяющие другие записи. Можно заметить, что дубликаты связаны только с идентификатором пользователя, поскольку если добавить в связку к `user_id` город банка, дубликатов не обнаруживается. Причиной может стать технический сбой, при которым разным пользователям в разных банках присваивается один идентификационный номер. Чтобы устранить дубликаты по пользователям, присвоим им новый идентификатор. Для этого вычислим минимальный идентификационный номер и каждый дубликат перезапишем через новую нумерацию, в которой каждый номер будет меньше другого на единицу.

```python
initial_min = df['user_id'].min()
initial_min
```

```python
dups_index = df[df.duplicated(subset=['user_id'])].index
print('Количество дубликатов:', len(dups_index))
```

Перезапишем идентификаторы записей дубликатов, используя новую нумерацию, описанную выше. Выведем новый минимальный идентификационный номер.

```python
df.loc[dups_index, 'user_id'] = range(df['user_id'].min() - 1, df['user_id'].min() - 74, -1)
print('Минимальный идентификационный номер:', df['user_id'].min())
```

Выведем все записи, в которых `user_id` меньше исходного минимального идентификатора - `initial_min`:

```python
df[df['user_id'] < initial_min]
```

```python
print('Количество дубликатов по идентификатору:', len(df[df.duplicated(subset='user_id')]))
```

Удалим пропуски в колонке с возрастом клиентов, в датафрейме их содержится меньше 1%, поэтому удаление не повлечет за собой сильной утери данных.

### Количественные данные

Функция для визуализации количественных данных - построение гистограммы действующих клиентов и клиентов в оттоке с единой осью частот - осью ординат и график плотности вероятности, отображающий распределения клиентов в оттоке и действующих относительно выбранного количественного признака.

```python
def plot_numeric_data(data, feature, bins, title_hist, title_density, xlabel, scaler, is_scaler=False):
    
    colors=['#07B1B1', '#ff5576']
    characteristic = ['действующих клиентов', 'клиентов в оттоке']
    fig, ax = plt.subplots(1, 3, figsize=(20, 7), sharey=True)
    
    if is_scaler:
        data.loc[:][feature] = data[feature] / scaler
        
    for i in range(2):
        ax[i].hist(data[data['churn'] == i][feature], bins=bins, color=colors[i])
        ax[i].set_title(f'{title_hist}\n{characteristic[i]}', fontsize=16)
        ax[i].set_xlabel(xlabel, fontsize=14)
        ax[i].set_ylabel('Частоты', fontsize=14)
        
    ax = plt.subplot(1, 3, 3)
    sns.histplot(ax=ax, data=data, hue='churn', x=feature, bins=bins, alpha=0.6, 
                 stat='density', common_norm=False, kde=True, palette=colors)
    plt.title(f'{title_density} \n', fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel('Плотность', fontsize=14)
    plt.legend(['Клиенты в оттоке', 'Действующие клиенты'], fontsize=13)
    
    plt.tight_layout()
```

Наибольшие значения оттока находятся в диапазоне 850-900 баллов кредитного скоринга. ### Категориальные данные

Функция для визуализации категориальных данных.

```python
def plot_categorical_data(df, feature, title_hist, title_density, xticks=False, bar_range=0):
    
    sort_df = df.sort_values(by=feature, ascending=False)
    plt.figure(figsize=(15, 6))
    colors=['#07B1B1', '#ff5576']

    ax = plt.subplot(1, 2, 1)
    sns.histplot(ax=ax, data=sort_df, hue='churn', x=feature, discrete=True, multiple="stack", shrink=0.7, 
                 alpha=0.8, palette=colors)
    plt.title(title_hist, fontsize=15)
    plt.xlabel('')
    plt.xticks(fontsize=14)
    plt.ylabel('Частоты', fontsize=12)
    plt.legend(['Клиенты в оттоке', 'Действующие клиенты'])
    if xticks:
        plt.xticks(bar_range)

    ax = plt.subplot(1, 2, 2)
    sns.histplot(ax=ax, data=sort_df, hue='churn', x=feature, discrete=True, stat='density', common_norm=False,
                 multiple="dodge", shrink=0.7, alpha=0.8, palette=colors)
    plt.title(title_density, fontsize=15)
    plt.xlabel('')
    plt.xticks(fontsize=14)
    plt.ylabel('Плотность', fontsize=12)
    plt.legend(['Клиенты в оттоке', 'Действующие клиенты'])
    if xticks:
        plt.xticks(bar_range)
```

Для каждого категориального признака посчитаем отток клиентов в разрезе всех уникальных категорий признака. Выводить уровень оттока будем только для тех категорий, которые превышают средний отток по всем клиентам.

Зафиксируем отток среди всех клиентов датафрейма:

```python
mean_churn_df = df['churn'].mean()
print('Доля оттока в датафрейме:', round(mean_churn_df, 3))
```

```python
def find_churn_categorical_data(df, features):
    
    for feature in features:
        print('Признак:', feature)
        uniq = df[feature].sort_values().unique()
        
        for el in uniq:
            churn = df[df[feature] == el]['churn'].mean().round(2)
            if churn >= mean_churn_df:
                print(f'Отток клиентов {el}: {churn}') 
        print()
```

```python
find_churn_categorical_data(df, features=['city', 'gender', 'equity', 'products', 'credit_card', 'last_activity'])
```

## Проверка гипотез

В ходе проверки гипотез будем рассматривать выборки, не зависящие друг от друга. Для проверки гипотез будем использовать t-тест.

### Гипотеза о равенстве доходов

Проверим, есть ли различия между доходами клиентов в оттоке и действующими клиентами:
    
**H<sub>0</sub>** - доходы клиентов в оттоке равняются доходам действующих клиентов;

**H<sub>1</sub>** - доходы клиентов в оттоке не равняются доходам действующих клиентов.

```python
alpha = 0.05

results = stats.ttest_ind(df[df['churn'] == 0]['est_salary'],
                          df[df['churn'] == 1]['est_salary'])

print('p-значение:', round(results.pvalue, 4))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

```python
print('Средний доход действующих клиентов:', round(df[df['churn'] == 0]['est_salary'].mean(), 2), 'руб.')
print('Средний доход клиентов в оттоке:', round(df[df['churn'] == 1]['est_salary'].mean(), 2), 'руб.')
```

```python
print('Медианный доход действующих клиентов:', round(df[df['churn'] == 0]['est_salary'].median(), 2), 'руб.')
print('Медианный доход клиентов в оттоке:', round(df[df['churn'] == 1]['est_salary'].median(), 2), 'руб.')
```

Таким образом, вывод о различии доходов клиентов в оттоке и действующих клиентов сделать нельзя. Средний доход между группа отличается менее чем на тысячу рублей.

### Гипотеза о равенстве среднего возраста

Проверим, есть ли различия в возрасте клиентов в оттоке и действующих клиентов:
    
**H<sub>0</sub>** - возраст клиентов в оттоке равняется возрасту действующих клиентов;

**H<sub>1</sub>** - возраст клиентов в оттоке не равняются возрасту действующих клиентов.

```python
results = stats.ttest_ind(df[df['churn'] == 0]['age'],
                          df[df['churn'] == 1]['age'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

```python
print('Средний возраст действующих клиентов:', round(df[df['churn'] == 0]['age'].mean(), 1))
print('Средний возраст клиентов в оттоке:', round(df[df['churn'] == 1]['age'].mean(), 1))
```

Таким образом, возраст клиентов в оттоке и действующих клиентов отличается, и эта разница статистически значима. Средний возраст отличается примерно на 1,5 года.
